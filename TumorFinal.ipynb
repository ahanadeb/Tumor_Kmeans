```python
# kmeans function
def k_means(bnew, no_samples, wght):
    features=np.size(bnew,1);
    # standardising
    bnew=(bnew-np.mean(bnew,0))/np.std(bnew,0);
    #initialise weight
    
    print("clustering run",end='')
    for w in range(0,10):
        print("...", (w)*10,"%", end='')
        if(w!=0):
            print("..", costfinal, end='')
            if(costfinal==0):
                break;
        bnew=np.divide(bnew,wght);
        
        for v in range(0,100): 
            # random initialisation
            k=4;
            muK= np.zeros((k,features));
            for i in range(0,k):
                j=random.randint(1,no_samples-3);
                muK[i]=bnew[j];
        #         print(muK.shape);  





            # distance 
            for z in range(0,40):
                tevt=np.zeros((k,1));
                C=np.zeros((np.size(bnew,0),1));
                for i in range(0,np.size(bnew,0)):
                    for j in range(0,k):
                        tevt[j]=abs(bnew[i]-muK[j]).sum(axis=1);
                    C[i]=np.argmin(tevt);

                # cluster assignment
                newmUK=np.zeros((np.size(muK,0),np.size(muK,1)));
                dist=np.zeros((k,1));
                for i in range(0,np.size(bnew,0)):
                    newmUK[int(C.item((i,0)))]=newmUK[int(C.item((i,0)))]+bnew[i];
                    dist[int(C.item((i,0)))]=dist[int(C.item((i,0)))]+1;

                # what to do with invalid clusters    
                for i in range(0,k):
                    if(dist[i]==0):
                        dist[i]=1;    
                muK=newmUK/dist;
        # cost
            costfun=0;
            wt_to_add=np.zeros((1,features));
            for g in range(0,np.size(bnew,0)):
                costfun=costfun+int(abs(bnew[g]-muK[int(C.item((g,0)))]).sum(axis=1));
                wt_to_add=wt_to_add+abs(bnew[g]-muK[int(C.item((g,0)))])
            if(v==0):
                final_wt_to_add=wt_to_add/no_samples;
                costr=costfun;
                final_dist=dist;
                final_C=C;
            else:
                if(costfun<costr):
                    final_wt_to_add=wt_to_add/no_samples;
                    costr=costfun;
                    final_dist=dist;
                    final_C=C;
        alpha =.7; #.7 gives 97.59
        wght=wght+(alpha*final_wt_to_add);
        if(w==0):
            costfinal=costr;
            dist_final = final_dist;
            c_final=final_C;
    
        else:
            if(costr<costfinal):
                costfinal = costr;
                dist_final = final_dist;
                c_final=final_C;
        
        
    return c_final, dist_final, costfinal;
```

```python
def rand_index_score(clusters, classes):

    tp_plus_fp = comb(np.bincount(clusters), 2).sum()
    tp_plus_fn = comb(np.bincount(classes), 2).sum()
    A = np.c_[(clusters, classes)]
    tp = sum(comb(np.bincount(A[A[:, 0] == i, 1]), 2).sum()
             for i in set(clusters))
    fp = tp_plus_fp - tp
    fn = tp_plus_fn - tp
    tn = comb(len(A), 2) - tp - fp - fn
    return (tp + tn) / (tp + fp + fn + tn)
```
